<H1 style="text-align: center;">The Collatz Conjecture: An Unsolved Mystery 🤯</H1>

<p>Welcome to a journey into one of the simplest to state yet most notoriously difficult problems in mathematics: <b style="color:red;">the Collatz Conjecture.</b></p>

<p>Also known as the <b style="color:blue;background-color: lightblue;">$3n + 1$ problem</b>, the <b style="color:blue;background-color: lightblue;">$3n + 1$ algorithm</b>, or Hasse's algorithm. the conjecture is built on a very basic rule applied to any positive integer <b style="color:red;">n</b>:</p>

<ul>
    <li><b style="color:red;">If n is even,</b> divide it by two. ($\frac{n}{2}$)</li>
    <li><b style="color:red;">If n is odd,</b> multiply it by three and add one. ($3n+1$)</li>
</ul>

<p>The conjecture states that if you start with <b style="color:red;">any</b> positive integer and repeatedly apply these rules, you will <b style="color:red;">always</b> eventually reach the number <b style="color:red;">1</b>.</p>

<h2 style="text-align: center;">Example Sequence</h2>

<p>Let's start with the number <b style="color:red;">6</b>:</p>

<pre>
6 → 6 / 2 = <strong>3</strong>
3 → (3 × 3) + 1 = <strong>10</strong>
10 → 10 / 2 = <strong>5</strong>
5 → (5 × 3) + 1 = <strong>16</strong>
16 → 16 / 2 = <strong>8</strong>
8 → 8 / 2 = <strong>4</strong>
4 → 4 / 2 = <strong>2</strong>
2 → 2 / 2 = <strong>1</strong>
</pre>

<p>The sequence for the starting number <b style="color:red;">6</b> is $6 \rightarrow 3 \rightarrow 10 \rightarrow 5 \rightarrow 16 \rightarrow 8 \rightarrow 4 \rightarrow 2 \rightarrow 1$. The conjecture holds true! At least for the number 6 anyway.In the Collatz conjecture, a loop (or cycle) occurs if a sequence of numbers eventually returns to a value it has already passed through, causing the sequence to repeat infinitely. The only known cycle for positive integers is the trivial loop: $4 \rightarrow 2 \rightarrow 1$, and then back to 4. The conjecture itself posits that every positive integer will eventually enter this specific 4-2-1 cycle. A major part of the mathematical challenge is proving that no non-trivial loops exist—patterns of numbers that cycle indefinitely without ever reaching 1. While computers have checked numbers up to approximately $2^{68}$ without finding any other cycles, a formal proof remains one of the great's unsolved problems in mathematics.</p>

<h2 style="text-align: center;">Visualizing the Collatz Conjecture</h2>

<p>Despite its simplicity, the Collatz Conjecture generates sequences that can be wildly unpredictable, making visualization a popular and powerful tool for exploration.</p>

<h2 style="text-align: center;">Collatz Sequences as Graphs</h2>

<p>One of the most common ways to see the conjecture in action is by graphing the steps of a single sequence. These Collatz sequence graphs typically plot the step number on the *x*-axis against the value of the number in the sequence on the *y*-axis. They often demonstrate dramatic, seemingly random peaks before the sequence plunges down toward the final 4 → 2 → 1 loop.</p>


```python
import matplotlib.pyplot as plt

def get_collatz(n):
    sequence = [n]
    while n > 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

# Generate sequence
start_val = 7
values = get_collatz(start_val)
steps = list(range(len(values)))

# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(steps, values, marker='o', color='royalblue', linestyle='-', linewidth=2)

# Formatting the graph
plt.title(f"Collatz Sequence starting at {start_val}", fontsize=14)
plt.xlabel("Step Number", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(steps)  # Ensure every step is labeled on X-axis

plt.show()
```

<h2>The Collatz Tree 🌳</h2>

Perhaps the most comprehensive visualization is the Collatz Tree (also called the Reverse Collatz Tree or Collatz Graph). Instead of starting at a number and going down to 1, this tree is built backwards from the number 1.

If the current number *n* is reached by the 3n + 1 rule, the previous number was (n - 1) / 3. This only works if *n* is a specific type of number.
If the current number *n* is reached by the n / 2 rule, the previous number was 2n.
By applying these reverse operations, a branching structure emerges, where every positive integer appears exactly once. This structure visually represents the idea that every starting number is eventually connected to the number 1. The sheer complexity and recursive nature of this tree highlight why a simple rule can generate such a difficult-to-prove conjecture.

<h2>The Three Tower Types (Modulo 3)</h2>
     <p>We classify every odd number, which forms the base of a Doubling Tower, into three distinct categories based on its value modulo 3:</p>
            <ul>
                <li><strong>Type $6n+1$:</strong> Bases like $1, 7, 13, 19, \ldots$ ($\equiv 1 \pmod 3$)</li>
                <li><strong>Type $6n+3$:</strong> Bases like $3, 9, 15, 21, \ldots$ ($\equiv 0 \pmod 3$)</li>
                <li><strong>Type $6n+5$:</strong> Bases like $5, 11, 17, 23, \ldots$ ($\equiv 2 \pmod 3$)</li>
            </ul>
            
<h3>The Non-Branching $6n+3$ Towers</h3>
     <p>The $6n+3$ towers are unique because they can never be reached by the $3N+1$ rule from a previous odd number. Since the result of $3N+1$ is always $\equiv 1 \pmod 3$, and all numbers in a $6n+3$ tower are $\equiv 0 \pmod 3$ (multiples of 3), no connection is possible. <b>These towers only ever feed OUT to another tower.</b></p>

<h2>The Tower Connectors: Linking Odd Numbers</h2>
<p>The Collatz tree is formed by the $3n+1$ rule connecting the odd base ($N$) of one tower to a specific even number ($R$) inside another tower. This number $R$ must be of the form $6k+4$ and applications of the reverse odd operation, $\frac{R-1}{3}$, yields the original odd number $N$.</p>

<h3>Algebraic Conditions for Connection Points</h3>
<p>The connection points ($R$) only exist in towers whose bases are of type $6n+1$ or $6n+5$ and follow specific formulas where $p \ge 0$ is an integer representing the tower level above the base:</p>
            
<p>Because the number $R$ will be congruent to $\mathbf{1 \pmod 3}$ applications of the reverse odd operation, $\frac{R-1}{3}$, yields an odd number $2k+1$. Since $2 \equiv -1 \pmod 3$, multiplying by $2^k$ alternates the modulo 3 value:</p>
            
$$(6n+1)\cdot 2^{2p+2} = R$$

$$ (6n+5) \cdot 2^{2p+1} = R $$

<p>By finding which level $p$ of the tower satisfies these conditions, we establish a precise algebraic relationship for every **Odd-to-Odd Connection** in the entire Collatz Tree.</p>

$$ 8(2^{2p})n+\frac{1}{3}(2^{2p+2}-1) \rightarrow (6n+1)\cdot 2^{2p+2} \rightarrow 6n+1$$

$$ 4(2^{2p})n+\frac{1}{3}(5\times2^{2p+1}-1) \rightarrow (6n+5) \cdot 2^{2p+1} \rightarrow 6n+5$$

<p>Where p and n are integers $\ge 0$.

# Collatz Odd-to-Odd Transitions: Modular Summary

## 1. Overview
The Collatz odd-to-odd mapping f(x) = (3x+1)/2^n partitions the natural numbers into three functional categories based on their residue modulo 3. These transitions are structured differently depending on the modular resolution.

---

## 2. Transitions Modulo 6
Modulo 6 provides the cleanest view of the "Tower Categories." Odd numbers fall into three residues:

* 3 mod 6 (Category 0): These are "Source Towers" (multiples of 3). They have an in-degree of 0 in the odd-to-odd graph. They map to {1, 5} but are never mapped to.
* 1 mod 6 (Category 1): Relay towers that map to {1, 5}. This residue contains the stable 1-loop (1 -> 1).
* 5 mod 6 (Category 2): Relay towers that map to {1, 5}. These represent the high-traffic hubs of the system.

**Transition Rules:**
- (Any Odd) -> {1, 5} mod 6.
- The residue 3 mod 6 is strictly a source.

---

## 3. Transitions Modulo 12
Modulo 12 reveals the parity constraints of the division step (n).

* Source Residues: {3, 9} (Category 0).
* Relay Residues: {1, 5, 7, 11}.

**Mapping Logic:**
The destination depends on the number of divisions by 2 (n) required to reach an odd number:
- If n is EVEN: Result y is in {1, 7} (Category 1).
- If n is ODD: Result y is in {5, 11} (Category 2).

**Sample Transitions:**
- 1 mod 12 -> {1, 5, 7}
- 3 mod 12 -> {5, 11}
- 5 mod 12 -> {1, 7}
- 7 mod 12 -> {5, 11}
- 9 mod 12 -> {1, 7}
- 11 mod 12 -> {1, 5, 11}

---

## 4. Path Density & Flow
- Category 0 Density: 0% (Forward), 33.3% (Reverse Branching).
- Category 1 Density: 33.3% (Forward Trajectory).
- Category 2 Density: 66.7% (Forward Trajectory).

This bias suggests that trajectories spend twice as much time in Category 2 (5, 11 mod 12) than in Category 1 (1, 7 mod 12) while traversing the Collatz graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

def get_next_odd(x):
    """The standard odd-to-odd Collatz shortcut."""
    y = 3 * x + 1
    while y % 2 == 0:
        y //= 2
    return y

def get_modular_edges(mod, samples=100):
    """
    Finds all possible transitions between odd residues mod M.
    Samples several values for each residue to capture non-deterministic jumps.
    """
    edges = set()
    # We only care about odd residues
    odd_residues = [r for r in range(mod) if r % 2 != 0]

    for r in odd_residues:
        for k in range(samples):
            x = r + (k * mod)
            if x == 0: continue
            y = get_next_odd(x)
            edges.add((r, y % mod))
    return list(edges)

def plot_modular_graph(mod, ax, title):
    G = nx.DiGraph()
    edges = get_modular_edges(mod)
    G.add_edges_from(edges)

    # Define colors based on Category (residue mod 3)
    color_map = []
    for node in G.nodes():
        residue_mod_3 = node % 3
        if residue_mod_3 == 0:
            color_map.append('#ff4d4d') # Red: Category 0
        elif residue_mod_3 == 1:
            color_map.append('#4d94ff') # Blue: Category 1
        else:
            color_map.append('#4dff88') # Green: Category 2

    pos = nx.circular_layout(G)
    nx.draw(G, pos, ax=ax, with_labels=True, node_color=color_map,
            node_size=1000, font_weight='bold', arrowsize=20,
            connectionstyle='arc3, rad=0.1')
    ax.set_title(title, fontsize=14)

# Create the visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

plot_modular_graph(6, ax1, "Odd to Odd Transitions Modulo 6")
plot_modular_graph(12, ax2, "Odd to Odd Transitions Modulo 12")

plt.tight_layout()
plt.show()
```

# Summary of Reverse Collatz Tree & Modular Transition Equations

## 1. Tower Construction
Every positive integer can be represented as an odd base $b$ multiplied by a power of 2:
$$x = b \cdot 2^n, \quad b \in \{2k+1 \mid k \in \mathbb{N}_0\}$$

## 2. The Three Tower Categories (mod 3)
Towers are partitioned based on the residue of the odd base $b$:
- **Category 0 (Sources):** $b \equiv 0 \pmod 3$
- **Category 1 (Relays):** $b \equiv 1 \pmod 3$
- **Category 2 (Relays):** $b \equiv 2 \pmod 3$

## 3. The Odd-to-Odd Map (Forward)
For any odd integer $x$, the next odd base $y$ is found via:
$$y = \frac{3x+1}{2^n}$$
Where $n$ is the unique integer such that $y$ is odd.

### Transition Residue Logic:
The residue of $y \pmod 3$ is determined by the parity of the exponent $n$:
- If $n$ is even: $y \equiv 1 \pmod 3$ (Category 1)
- If $n$ is odd: $y \equiv 2 \pmod 3$ (Category 2)

## 4. The Predecessor Formulas (Reverse Map)
To find an odd predecessor $x$ from a base $b$, given a parameter $p \in \mathbb{N}_0$:

### For $b \in \text{Category 1 } (6k+1)$:
$$x = \frac{b \cdot 4^{p+1} - 1}{3}$$

### For $b \in \text{Category 2 } (6k+5)$:
$$x = \frac{b \cdot (2 \cdot 4^p) - 1}{3}$$

## 5. Statistical Densities
The probability $P$ of a forward trajectory landing in a specific category:
$$P(y \in \text{Cat 1}) = \sum_{k=1}^{\infty} (1/2)^{2k} = 1/3$$
$$P(y \in \text{Cat 2}) = \sum_{k=1}^{\infty} (1/2)^{2k-1} = 2/3$$

## 6. The Global Path Equation
The general form connecting a starting number $N$ to an ending number $M$ after $n$ odd steps and $m$ even steps:
$$M = \frac{3^n N + f}{2^m}$$
Where $f$ is the "path constant" resulting from the collected $+1$ residies from each odd step.


# Collatz Sequence Analysis Using Binary Tree structure

Every seed number N, as well as each number that results from applying the Collatz functions follows a unique trajectory that can be mapped onto an infinite binary tree. From any node N we travel down the left branch is N is even, and we travel down the right branch if N is odd:

A Right Turn represents an “Odd” step: $𝑁 →(3𝑁 +1)/2$

A Left Turn represents an “Even” step: $𝑁 →𝑁/2$

When we follow a specific path of n odd steps and m total steps, we aren’t just performing arithmetic; we are traversing a specific branch of this tree. This traversal generates a “Path Equation” that links the starting value N to the terminal value M:

$$\frac{3^𝑛𝑁 +𝑓}{2^𝑚} =𝑀$$

##2. How the Tree Generates the Constant f

​The constant f is the most fascinating part of this equation. It is not a static number but a “path memory” generated by the sequence of turns. The value for f is a result of the collected $+1$ residues from each odd step.

​If we define the path as a bit-string where 1 (or O) is an odd step and 0 (oe E) is an even step, f is built iteratively. Starting at f = 0 and m = 0:

​If the next step is Even (0): m increases by 1. f remains unchanged.

​If the next step is Odd (1): f becomes 3f + 2^m. Then n increases by 1 and m increases by 1.

​This means f is a weighted sum of powers of 2, where the weights are determined by the powers of 3 remaining in the path. The f variable is a sum of $3^n2^m$ terms.

Beginning with the path equation and a parity vector, like 101010 we can determine the values of n. m, and f. n is just the number of odd steps and m is the path length. f is determined using the positional formula:

$$∑^𝑛_{𝑖=1}3^{𝑛−𝑖}2^{𝑝_𝑖−1}$$

Where n is the number of odd steps, m is the total path length, and $𝑝_𝑖$ is the position of the ith odd step. In the case of the example path 101010, we get values for n, m, and f of $n = 3$, $m = 6$, and $f = 3^22^0+3^12^2+3^02^4$ or $f = 37$. This yields a path equation of:

$$\frac{3^3N+37}{2^6}=M$$

What the binary tree represents with respect to the Collatz sequence is possible paths given some seed value N. Imagine being handed a sealed envelope marked "N". Inside this envelope is a piece of paper with some positive integer greater than 1 written on it. That is the unknown seed value. With that limited information we can only generate a map of possible paths created when the Collatz functions are applied. Where N takes the left path if even with a probability of $\frac{1}{2}$ and it takes the right path if odd with a probability of $\frac{1}{2}$. When we open the envelope and N is revealed the tree collapses to a specific path down the tree, generating a unique path equation and specific values for N and M.

```python
import networkx as nx
import matplotlib.pyplot as plt

def build_collatz_tree(max_m):
    G = nx.DiGraph()
    # Initial root state: m=0, n=0, f=0, path string ""
    nodes = [("", 0, 0, 0)]
    G.add_node("", label="(root)\nn=0  f=0", level=0)

    for m in range(max_m):
        next_nodes = []
        for path, current_n, current_f, current_m in nodes:
            # 1. Left Child: E-step (Even)
            # Formula: n remains same, f remains same
            e_path = path + "E"
            G.add_node(e_path, label=f"{e_path}\nn={current_n}  f={current_f}", level=m+1)
            G.add_edge(path, e_path, type='E')
            next_nodes.append((e_path, current_n, current_f, m+1))

            # 2. Right Child: O-step (Odd)
            # Formula: n increases by 1, f becomes 3*f + 2^m
            o_path = path + "O"
            new_n = current_n + 1
            new_f = 3 * current_f + (2 ** current_m)
            G.add_node(o_path, label=f"{o_path}\nn={new_n}  f={new_f}", level=m+1)
            G.add_edge(path, o_path, type='O')
            next_nodes.append((o_path, new_n, new_f, m+1))

        nodes = next_nodes
    return G

def plot_tree(G):
    # Use a custom layout for binary tree structure
    pos = {}
    for node, data in G.nodes(data=True):
        level = data['level']
        # Calculate x position based on the binary path string
        path = node
        x = 0
        if path:
            # Map E to -1 and O to 1, scaled by level depth
            for i, char in enumerate(path):
                shift = 1 / (2 ** (i + 1))
                x += shift if char == "O" else -shift
        pos[node] = (x, -level)

    plt.figure(figsize=(20, 12))

    # Draw edges with colors matching your image
    e_edges = [(u, v) for u, v, d in G.edges(data=True) if d['type'] == 'E']
    o_edges = [(u, v) for u, v, d in G.edges(data=True) if d['type'] == 'O']

    nx.draw_networkx_edges(G, pos, edgelist=e_edges, edge_color='forestgreen', width=2)
    nx.draw_networkx_edges(G, pos, edgelist=o_edges, edge_color='firebrick', width=2)

    # Draw nodes as boxes
    labels = nx.get_node_attributes(G, 'label')
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=9, font_weight='bold',
                            bbox=dict(facecolor="#f0f0f0", edgecolor="black", boxstyle="round,pad=0.5"))

    # Add Legend and Titles
    plt.title("Collatz Binary Tree – E/O Path Labels", fontsize=18, pad=20)
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color='forestgreen', lw=2, label='E-step (even): N/2'),
        Line2D([0], [0], color='firebrick', lw=2, label='O-step (odd): (3N+1)/2')
    ]
    plt.legend(handles=legend_elements, loc='lower center', ncol=2, frameon=False, fontsize=12)
    plt.axis('off')
    plt.show()

# Generate tree for rows m = 0 to 3
collatz_tree = build_collatz_tree(3)
plot_tree(collatz_tree)
```

## The Path Equation as a Linear Diophantine Equation

Every trajectory in the Collatz conjecture can be encoded as a Path Vector (a bit-string where $1$ represents an odd step $\frac{3n+1}{2}$ and $0$ represents an even step $\frac{n}{2}$. This vector represents a specific path down the binary tree.  Based on the method outline above we can determine the values of n, m, and f from the path vector. We can then rearrange the path equation into the form of a Linear Diophantine Equation:

$$2^m M - 3^n N = f$$

In this framework:

* $N$: The starting integer (Seed).
* $M$: The resulting integer after the path is completed.
* $n$: Total count of odd steps (weight of the $3$ multiplier).
* $m$: The path length as well as the total count of divisions by 2.
* $f$: The Path Constant, representing the accumulated "residue" of the $+1$ shifts throughout the trajectory.

By calculating these parameters from a given path vector, we can use the Extended Euclidean Algorithm to solve for $N$ and $M$. This allows us to determine exactly which integers $N$ are capable of following a specific path vector.


```python
def solve_collatz_path(path_vector):
    """
    Solves for n, m, f, N, and M given a path vector.
    1 = Odd step (3n+1)/2, 0 = Even step (n/2)
    """
    n = 0  # Number of odd steps
    m = 0  # path length
    f = 0  # Path memory constant

    # Process the path vector in reverse to build the f constant
    # Equation: M = (3^n * N + f) / 2^m
    current_3_power = 0

    for step in path_vector:
        if step == 1:
            # Odd step: (3N + 1)/2
            # f tracks the accumulation of '+1' multiplied by subsequent 3s
            f += 3**current_3_power * 2**m
            n += 1
            m += 1
            current_3_power += 1
        else:
            # Even step: N / 2
            m += 1

    # Solve the Diophantine Equation: 2^m * M - 3^n * N = f
    # We use the modular inverse to find N
    # N = (-f * inv(3^n, 2^m)) mod 2^m
    try:
        # Calculate N using modular arithmetic
        # We need (3^n * N) % 2^m == (-f) % 2^m
        inv_3n = pow(3**n, -1, 2**m)
        N = ((-f) * inv_3n) % (2**m)

        # If N is 0, the smallest positive N is 2^m
        if N == 0: N = 2**m

        # Calculate M based on the derived N
        M = (3**n * N + f) // (2**m)

        return {
            "n": n,
            "m": m,
            "f": f,
            "N_min": N,
            "M_result": M
        }
    except ValueError:
        return "No modular inverse exists (this shouldn't happen with powers of 2 and 3)."

# Example: Path 1 -> 4 -> 2 -> 1
# This is: Odd (1), Even (0), Even (0)
vector = [1, 0, 0, 0]
results = solve_collatz_path(vector)

print(f"Path Vector: {vector}")
print(f"Odd Steps (n): {results['n']}")
print(f"Path Length (m): {results['m']}")
print(f"Path Constant (f): {results['f']}")
print(f"Smallest Seed (N): {results['N_min']}")
print(f"Terminal Value (M): {results['M_result']}")
```
